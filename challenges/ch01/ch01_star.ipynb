```python
# Revenue by day — no joins needed
ch.db.query('''
    SELECT date_key, day_of_week, is_payday_week,
           COUNT(DISTINCT transaction_id) AS txn_count,
           SUM(quantity_sold * unit_price_thb) AS revenue,
           COUNT(DISTINCT customer_id) AS customers
    FROM fact_sales
    GROUP BY date_key
    ORDER BY date_key
    LIMIT 15
''')
```

```python
# Revenue by category
ch.db.query('''
    SELECT p.category, COUNT(*) AS items_sold,
           SUM(fs.quantity_sold * fs.unit_price_thb) AS revenue
    FROM fact_sales fs
    JOIN dim_product p ON fs.product_key = p.product_key
    GROUP BY p.category
    ORDER BY revenue DESC
''')
```

```python
# Inventory — low stock
ch.db.query('''
    SELECT fi.date_key, p.product_name,
           fi.opening_stock, fi.units_sold, fi.closing_stock
    FROM fact_inventory fi
    JOIN dim_product p ON fi.product_key = p.product_key
    WHERE fi.closing_stock < 10
    ORDER BY fi.closing_stock
    LIMIT 10
''')
```
