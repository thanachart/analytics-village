```python
# Transactions — one row per item sold
ch.db.query('SELECT * FROM transactions LIMIT 10')
```

```python
# Join transactions with products and calendar
ch.db.query('''
    SELECT t.transaction_date, t.customer_id,
           p.product_name, p.category,
           t.quantity_sold, t.unit_price_thb,
           c.day_of_week, c.is_payday_week
    FROM transactions t
    JOIN products p ON t.product_id = p.product_id
    JOIN calendar c ON t.transaction_date = c.date
    LIMIT 10
''')
```

```python
# Daily revenue (join with calendar)
ch.db.query('''
    SELECT c.date, c.day_of_week, c.is_payday_week, c.event_name,
           COUNT(DISTINCT t.transaction_id) AS txn_count,
           SUM(t.quantity_sold * t.unit_price_thb) AS revenue
    FROM calendar c
    LEFT JOIN transactions t ON c.date = t.transaction_date
    GROUP BY c.date
    ORDER BY c.date
    LIMIT 15
''')
```

```python
# Inventory — low stock days
ch.db.query('''
    SELECT i.record_date, p.product_name, p.category,
           i.opening_stock, i.units_sold, i.units_received, i.closing_stock
    FROM inventory i
    JOIN products p ON i.product_id = p.product_id
    WHERE i.closing_stock < 30 AND i.record_date != ''
    ORDER BY i.closing_stock
    LIMIT 10
''')
```
