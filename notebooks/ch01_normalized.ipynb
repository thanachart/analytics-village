```python
# Transactions + items (JOIN practice)
ch.db.query('''
    SELECT t.transaction_id, t.transaction_date, t.customer_id,
           t.total_amount_thb, t.satisfaction,
           ti.product_id, p.product_name, p.category,
           ti.quantity_sold, ti.unit_price_thb, ti.line_total_thb,
           ti.had_stockout
    FROM transactions t
    JOIN transaction_items ti ON t.transaction_id = ti.transaction_id
    JOIN products p ON ti.product_id = p.product_id
    LIMIT 10
''')
```

```python
# Daily revenue (join transactions with calendar for context)
ch.db.query('''
    SELECT c.date, c.day_of_week, c.is_payday_week, c.event_name,
           COUNT(DISTINCT t.transaction_id) AS txn_count,
           SUM(t.total_amount_thb) AS revenue
    FROM calendar c
    LEFT JOIN transactions t ON c.date = t.transaction_date
    GROUP BY c.date
    ORDER BY c.date
    LIMIT 15
''')
```

```python
# Inventory snapshot
ch.db.query('''
    SELECT i.record_date, p.product_name, i.opening_stock,
           i.units_sold, i.units_received, i.closing_stock, i.had_stockout
    FROM inventory i
    JOIN products p ON i.product_id = p.product_id
    WHERE i.had_stockout = 1
    ORDER BY i.record_date
    LIMIT 10
''')
```
