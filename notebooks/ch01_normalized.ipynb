```python
# Transaction header + items (practice your JOINs!)
ch.db.query('''
    SELECT t.transaction_id, t.transaction_date, t.customer_id,
           t.total_amount_thb, t.satisfaction,
           ti.product_id, p.product_name, ti.quantity_sold, ti.line_total_thb
    FROM transactions t
    JOIN transaction_items ti ON t.transaction_id = ti.transaction_id
    JOIN products p ON ti.product_id = p.product_id
    LIMIT 10
''')
```

```python
# Daily revenue with calendar context
ch.db.query('''
    SELECT c.date, c.day_of_week, c.is_payday_week, c.event_name,
           COUNT(t.transaction_id) AS txn_count,
           SUM(t.total_amount_thb) AS revenue
    FROM calendar c
    LEFT JOIN transactions t ON c.date = t.transaction_date
    GROUP BY c.date
    ORDER BY c.date
    LIMIT 15
''')
```

```python
# Customer lifecycle — who churned?
ch.db.query('''
    SELECT cl.customer_id, cl.event_date, cl.previous_state, cl.new_state,
           cl.trigger_reason, cl.days_since_last_visit,
           cust.income_bracket, cust.household_size
    FROM customer_lifecycle cl
    JOIN customers cust ON cl.customer_id = cust.customer_id
    ORDER BY cl.event_date DESC
    LIMIT 10
''')
```
