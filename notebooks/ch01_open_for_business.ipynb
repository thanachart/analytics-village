### Submit via GitHub

Run the cell below to automatically commit and push your submission.

Or manually:
1. Go to https://github.com/thanachart/analytics-village
2. Navigate to `submissions/ch01/`
3. Upload your JSON file
4. Create a Pull Request

```python
import subprocess, os, json

def submit_to_github(filepath, challenge_id='ch01', repo_url='https://github.com/thanachart/analytics-village'):
    """Submit your decision.json to the GitHub repository."""
    if not os.path.exists(filepath):
        print(f'Error: File not found: {filepath}')
        return
    
    with open(filepath) as f:
        data = json.load(f)
    student_id = data.get('metadata', {}).get('student_ids', ['unknown'])[0]
    filename = os.path.basename(filepath)
    
    repo_root = os.path.abspath(os.path.join(os.path.dirname(filepath), '..', '..'))
    branch_name = f'submit/{challenge_id}/{student_id}'
    
    print(f'Submitting {filename} to {repo_url}')
    print(f'Branch: {branch_name}')
    
    cmds = [
        f'cd "{repo_root}"',
        f'git checkout -b {branch_name} 2>/dev/null || git checkout {branch_name}',
        f'git add submissions/{challenge_id}/{filename}',
        f'git commit -m "Submit {challenge_id} — {student_id}"',
        f'git push origin {branch_name}',
    ]
    
    for cmd in cmds:
        print(f'  > {cmd}')
        result = subprocess.run(cmd, shell=True, capture_output=True, text=True, cwd=repo_root)
        if result.returncode != 0 and result.stderr:
            print(f'    {result.stderr.strip()}')
    
    print(f'\nDone! Create a Pull Request at:')
    print(f'  {repo_url}/compare/main...{branch_name}')

# Run this to submit:
# submit_to_github(filepath)
```
